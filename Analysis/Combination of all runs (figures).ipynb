```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
no_GT_logtimes = [-1e9,6.698,7.0,7.176,7.301,7.397,7.477,7.544,7.602,7.653,7.698,7.740,7.778,7.812,7.845,7.875,7.903,7.929,7.954,7.977]
no_GT_times = 10**np.array(no_GT_logtimes)

GT_logtimes = [-1e9,4.300,4.602,4.778,4.903,4.999,5.0,5.002,5.004,5.006,5.008,5.010,5.305,5.479,5.603,5.699,5.778,6.745,7.022,7.190,7.311,7.405,7.483,7.548,7.605,7.656,7.701,7.742,7.779,7.814,7.846,7.875,7.903,7.929,7.954,7.977]
GT_times = 10**np.array(GT_logtimes)
```

# Potential Oort percentage

```python
pot_Oort_Van_no_GT = np.load('npy files/pot_Oort_no_GT.npy')
pot_Oort_Van_GT = np.load('npy files/pot_Oort_GT.npy')
pot_Oort_MWG_no_GT = np.load('npy files/pot_Oort_MWG_no_GT.npy')
pot_Oort_MWG_GT = np.load('npy files/pot_Oort_MWG_GT.npy')
```

```python
amount_of_runs = [len(pot_Oort_Van_no_GT), len(pot_Oort_Van_GT), len(pot_Oort_MWG_no_GT), len(pot_Oort_MWG_GT)]
```

```python
plt.errorbar(no_GT_times/1e6, np.mean(pot_Oort_Van_no_GT, axis=0)*100, np.std(pot_Oort_Van_no_GT, axis=0)/np.sqrt(amount_of_runs[0])*100, capsize=4, label='Vanilla')
plt.errorbar(GT_times/1e6, np.mean(pot_Oort_Van_GT, axis=0)*100, np.std(pot_Oort_Van_GT, axis=0)/np.sqrt(amount_of_runs[1])*100, capsize=4, label='Vanilla Grand Tack')
plt.errorbar(no_GT_times/1e6, np.mean(pot_Oort_MWG_no_GT, axis=0)*100, np.std(pot_Oort_MWG_no_GT, axis=0)/np.sqrt(amount_of_runs[2])*100, capsize=4, label='MWG')
plt.errorbar(GT_times/1e6, np.mean(pot_Oort_MWG_GT, axis=0)*100, np.std(pot_Oort_MWG_GT, axis=0)/np.sqrt(amount_of_runs[3])*100, capsize=4, label='MWG Grand Tack')


plt.xlabel("Time [Myr]")
plt.ylabel("Percentage of potential Oort particles [%]")
plt.grid()
plt.ylim(-0.1, 5.6)
plt.xlim(-2, 100)
plt.title('Percentage of potential Oort particles in time')
plt.legend(title='Model')
plt.show()
plt.savefig('Figures/pot_Oort.png')
```

# Perturbed percentage

```python
perturbed_Van_no_GT = np.load('npy files/perturbed_no_GT.npy')
perturbed_Van_GT = np.load('npy files/perturbed_GT.npy')
perturbed_MWG_no_GT = np.load('npy files/perturbed_MWG_no_GT.npy')
perturbed_MWG_GT = np.load('npy files/perturbed_MWG_GT.npy')
```

```python
plt.errorbar(no_GT_times/1e6, np.mean(perturbed_Van_no_GT, axis=0)*100, np.std(perturbed_Van_no_GT, axis=0)/np.sqrt(amount_of_runs[0])*100, capsize=4, label='Vanilla')
plt.errorbar(GT_times/1e6, np.mean(perturbed_Van_GT, axis=0)*100, np.std(perturbed_Van_GT, axis=0)/np.sqrt(amount_of_runs[1])*100, capsize=4, label='Vanilla Grand Tack')
plt.errorbar(no_GT_times/1e6, np.mean(perturbed_MWG_no_GT, axis=0)*100, np.std(perturbed_MWG_no_GT, axis=0)/np.sqrt(amount_of_runs[2])*100, capsize=4, label='MWG')
plt.errorbar(GT_times/1e6, np.mean(perturbed_MWG_GT, axis=0)*100, np.std(perturbed_MWG_GT, axis=0)/np.sqrt(amount_of_runs[3])*100, capsize=4, label='MWG Grand Tack')

plt.xlabel("Time [Myr]")
plt.ylabel("Percentage of perturbed particles [%]")
plt.grid()
plt.ylim(0, 50)
plt.xlim(-2, 100)
plt.title('Percentage of perturbed particles (a > 100 au)')
plt.legend(title='Model')
plt.show()
plt.savefig('Figures/perturbed.png')
```

# Scattering rate

```python
scattering_rate_Van_no_GT = np.load('npy files/scattering_rate_no_GT.npy')
scattering_rate_Van_GT = np.load('npy files/scattering_rate_GT.npy')
scattering_rate_MWG_no_GT = np.load('npy files/scattering_rate_MWG_no_GT.npy')
scattering_rate_MWG_GT = np.load('npy files/scattering_rate_MWG_GT.npy')

scattering_rate_Van_no_GT_std = np.load('npy files/scattering_rate_no_GT_std.npy')
scattering_rate_Van_GT_std = np.load('npy files/scattering_rate_GT_std.npy')
scattering_rate_MWG_no_GT_std = np.load('npy files/scattering_rate_MWG_no_GT_std.npy')
scattering_rate_MWG_GT_std = np.load('npy files/scattering_rate_MWG_GT_std.npy')
```

```python
#plt.figure(figsize=(8,6))

plt.errorbar(no_GT_times[1:]/1e6, scattering_rate_Van_no_GT, scattering_rate_Van_no_GT_std, capsize=4, marker='.', linestyle='None', label='Vanilla')
plt.errorbar(GT_times[1:]/1e6, scattering_rate_Van_GT, scattering_rate_Van_GT_std, capsize=4, marker='.', linestyle='None', label='Vanilla Grand Tack')
plt.errorbar(no_GT_times[1:]/1e6, scattering_rate_MWG_no_GT, scattering_rate_MWG_no_GT_std, capsize=4, marker='.', linestyle='None', label='MWG')
plt.errorbar(GT_times[1:]/1e6, scattering_rate_MWG_GT, scattering_rate_MWG_GT_std, capsize=4, marker='.', linestyle='None', label='MWG Grand Tack')

plt.vlines(0.1025, -10, 100, linestyle='--', label='Inward/Outward migration', color='black')
plt.vlines(0.6, -10, 100, linestyle='-.', label='End of Outward migration', color='black', zorder=-2)

plt.legend(title='Model')

plt.plot(no_GT_times[1:]/1e6, scattering_rate_Van_no_GT, marker='.', alpha=0.2, label='Vanilla', color='C0')
plt.plot(GT_times[1:]/1e6, scattering_rate_Van_GT, marker='.', alpha=0.2, label='Vanilla Grand Tack', color='C1')
plt.plot(no_GT_times[1:]/1e6, scattering_rate_MWG_no_GT, marker='.', alpha=0.2, label='MWG', color='C2')
plt.plot(GT_times[1:]/1e6, scattering_rate_MWG_GT, marker='.', alpha=0.2, label='MWG Grand Tack', color='C3')


plt.xscale('log')
plt.ylim(-1, 25)
plt.grid()
plt.ylabel('Ejection rate [% / Myr]')
plt.xlabel('Time [Myr]')
plt.title(r'Ejection rate (percentage of ejected particles divided by $\Delta t$)')
plt.show()
plt.savefig('Figures/scattering_rate.png')
```
